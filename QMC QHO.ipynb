# Quantum Monte-Carlo method

## Theory:

1.  In quantum mechanics
 we can express the expectation value of a given operator O for a system of N particles as

 $\langle \hat{O} \rangle = \frac{\int d\mathbf{R}_1 d\mathbf{R}_2 \dots d\mathbf{R}_N 
 \Psi^*(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N) 
 \hat{O}(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N) 
 \Psi(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N)}
 {\int d\mathbf{R}_1 d\mathbf{R}_2 \dots d\mathbf{R}_N 
 \Psi^*(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N) 
 \Psi(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N)}$,
 where $\Psi(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N)$ is the wave function describing a many-body system.

2.  We can write Schrodinger's equation as a differential equation with the energy operator $\hat{H}$ i.e., the Hamiltonian operator of the system acting on the wavefunction describing the system, as: 

$\hat{H}\Psi(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N, \alpha_A, \alpha_B, \dots, \alpha_N) = E\Psi(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N, \alpha_A, \alpha_B, \dots, \alpha_N)$

where $\mathbf{R}_1,\mathbf{R}_2,....\mathbf{R}_A$ are the coordinates of the particles of the many body system and $\alpha_A,\alpha_B, \dots, \alpha_N$ are the set of relevant quantum numbers such as spin and isospin of a system(let's say)

3. The above equation is a multi-dimensional integral and **Monte Carlo methods** are ideal for obtaining expectation values of such quantum mechanical operators. 

4. The problem lies in the fact that most of the time we do not the exact wavefunction $\Psi(\mathbf{R}_1, \mathbf{R}_2, \dots, \mathbf{R}_N, \alpha_A, \alpha_B, \dots, \alpha_N)$.

5. We can try introducing a function which depends on selected variational parameters. This function should essentially capture the features of the system under consideration.

6. With such a wavefunction, we will therefore try to perform a variational calculation of various observables using Monte-Carlo methods.

## Some postulates of quantum mechanics:

1. $i\hbar \frac{\partial}{\partial t} \Psi(\mathbf{r}, t) = 
\left[ -\frac{\hbar^2}{2m} \nabla^2 + V(\mathbf{r}, t) \right] \Psi(\mathbf{r}, t)$ 

is the time dependent schrodinger equation. $V(\mathbf{r},t)$ is the potential acting on the particle in the system described by the wavefunction $\Psi$ and the first term on RHS is the kinetic energy term. The solution to this partial differential equation is $\Psi(\mathbf{r},t)$

2. The quantum mechanical probability to find a particle anywhere in $(\mathbf{x},\mathbf{x+dx})$  is given by $P(\mathbf{x},t) \, dx = \Psi^{*}(\mathbf{x},t) \Psi(\mathbf{x},t) \, dx$

3. Wavefunction of a system is a complex function, having both real and imaginary side. So, the $\Psi^{*}(\mathbf{x},t) \Psi(\mathbf{x},t)$ = $R^2 + I^2$ where, $\Psi(\mathbf{x},t) = R(\mathbf{x},t)+\iota I(\mathbf{x},t)$. 

4. **The variational monte carlo approach uses this definition of probability, allowing us thereby to deal with real-quantities only.**

## The Born Interpretation and Conditions on $( \Psi(x,t) )$

The Born interpretation constrains the wave function to belong to the class of functions in $( L^2 )$. Some of the selected conditions that $( \Psi(x,t) )$ must satisfy are:

1. **Normalization**:  
   Meaning that  
   $$
   \int_{-\infty}^{\infty} P(x,t) \,dx = \int_{-\infty}^{\infty} \Psi^*(x,t) \Psi(x,t) \,dx = 1.
   $$

2. $( \Psi(x,t) )$ and $( \frac{\partial \Psi(x,t)}{\partial x} )$ must be finite.

3. $( \Psi(x,t) )$ and $( \frac{\partial \Psi(x,t)}{\partial x} )$ must be continuous.

4. The function must be square-integrable:
   $$
   \int_{-\infty}^{\infty} \Psi^*(x,t) \Psi(x,t) \,dx < \infty.
   $$

5. $( \Psi(x,t) )$ and $( \frac{\partial \Psi(x,t)}{\partial x} )$ must be single-valued.


## The variational monte carlo method:

To begin with the variational monte-carlo method, we should first choose an appropriate wavefunction that encapsulates the details of a system. These wavefunctions establishes the starting point of variational calculation of the expectation value of the Hamiltonian H.

Now, given a hamiltonian $\hat{H}$ and a trial wave function $\Psi_{T}$, the variational principle states that the expectation value of $\langle \hat{H} \rangle$ is:

$H = \frac{\int dR \, \Psi_{T}^*(R) H(R) \Psi_T(R)}
{\int dR \, \Psi^*(R) T(R) \Psi_T(R)},$

which serves as an upper bound to the true ground state energy $( E_0 )$ of the Hamiltonian $( \hat{H} )$, meaning that:

$E_0 \leq \langle H \rangle$

To show this, we note first that the trial wave function can be expanded in the eigenstates of the Hamiltonian since they form a complete set, see again Postulate III,  

$ \Psi_T(\mathbf{R}) = \sum_i a_i \Psi_i(\mathbf{R}) $,  

and assuming the set of eigenfunctions to be normalized, insertion of the latter equation in the above equation results in  

$ \langle H \rangle = \frac{\sum_{mn} a_m^* a_n \int d\mathbf{R} \Psi_m^*(\mathbf{R}) H(\mathbf{R}) \Psi_n(\mathbf{R})}
{\sum_{mn} a_m^* a_n \int d\mathbf{R} \Psi_m^*(\mathbf{R}) \Psi_n(\mathbf{R})} $  

$ = \frac{\sum_{mn} a_m^* a_n \int d\mathbf{R} \Psi_m^*(\mathbf{R}) E_n(\mathbf{R}) \Psi_n(\mathbf{R})}
{\sum_n a_n^2} $,  

which can be rewritten as  

$ \frac{\sum_n a_n^2 E_n}{\sum_n a_n^2} \geq E_0 $.


In general, the integrals involved in the calculation of various expectation values are multi-dimensional
 ones. Traditional integration methods such as the Gauss-Legendre will not be adequate for say the
 computation of the energy of a many-body system. The fact that we need to sample over a multi
dimensional density and that the probability density is to be normalized by the division of the norm of
 the wave function, **suggests that e.g., the Metropolis algorithm may be appropriate.**

 ### Steps to perform variational monte carlo method:
1. Construct a trial wave function first $\Psi_T(\mathbf{R};\alpha)$ for a many body system consisting of N-particles locates at positions: $\mathbf{R} = (\mathbf{R}_1,\mathbf{R}_2,\mathbf{R}_3, \dots, \mathbf{R}_N)$. The trial wavefunction also depends on $\alpha$ variational parameters $\alpha = \alpha_1,\alpha_2,\dots,\alpha_N$
2. Evaluate the expectation value of Hamiltonian $\hat{H}$
3. Then vary $\alpha$ according to some minimization algorithm and return to the first step.

The above loop stops when we reach the minimum of the energy according to some specified criterion. 

**Note:** 

1. In most cases, a wave function has only small values in large parts of configurationspace, and a straight
forward procedurewhich uses homogenouslydistributed randompoints in configurationspace will most
likely lead to poor results. This may suggest that some kind of importance sampling combined with e.g.,
the Metropolis algorithm may be a more efficient way of obtaining the ground state energy. The hope is
then that those regions of configurations space where the wave function assumes appreciable values are
sampled more efficiently.

2. The tedious part in a variational Monte Carlo calculation is the search for the variational minimum. A good knowledge of the system is required in order to carry out reasonable variational Monte Carlo
calculations. 


# Variational Monte-Carlo methods for quantum mechanical systems:

The **Variational Monte-Carlo method** is a numerical technique used to approximate the ground state energy of a quantum system. The idea is based on the variational principle which states that any **trial wave function** $\Psi_{T}(\mathbf{R};\alpha)$, the expectation value of the hamiltonian provides an upper bound to the exact ground state energy of the system under study.

$\mathbf{R}$: represents the spatial co-ordinates of the system.

$\alpha$: is a variational parameter **used to optimize the wavefunction.**

## Probability distribution in VMC:

To efficiently sample quantum states, we will define a probability distribution function based on the square magnitude of the trial wave function:

$P(\mathbf{R};\alpha) = \frac{|\Psi_{T}(\mathbf{R};\alpha)|^{2}}{\int |\Psi_{T}(\mathbf{R};\alpha)|^{2} dR }$

This distribution will be used for sampling in monte-carlo integration. Instead of sampling all possibe quantum states, we will prioritize the states where $\Psi_{T}(\mathbf{R};\alpha)$ is large.

## The local energy-operator:

**The local energy-operator is defined as:**

$\hat{E}_L(\mathbf{R}; \alpha) = \frac{1}{\psi_T(\mathbf{R}; \alpha)} \hat{H} \psi_T(\mathbf{R}; \alpha)$

This quantity represents the energy at a given configuration $\mathbf{R}$. Using the probability distribution $P(\mathbf{R};\alpha)$, we compute the expectation value of the local energy:

$\langle E_{L}(\alpha) \rangle = \int P(\mathbf{R};\alpha)\hat{E_{L}}(\mathbf{R};\alpha)d\mathbf{R}$

**Note:** If the trial wavefunction $\Psi_T$ is close to the exact wavefunction, then $\langle E_L(\alpha) \rangle$ should approximate the exact energy $\hat{H}$. 

**The goal of VMC is to optimize the parameter $\alpha$ to minimize $\langle E_L(\alpha) \rangle$**, since the variational principle ensures that this gives the best approximation to the ground state energy

# The algorithm:

 The numerically time-consuming part in the variational Monte Carlo calculation is the evaluation of the
 kinetic energy term. The potential energy, as long as it has a spatial dependenceonly, adds a simple term
 to the local energy operator

  In our discussion below, we base our numerical Monte Carlo solution on the Metropolis algorithm.
 **The implementation is rather similar to the one discussed in connection with the Ising model, the main
 difference resides in the form of the PDF.** The main test to be performed by the Metropolis algorithm is
 a ratio of probabilities.

 We need to perform the following 2 states:

 1. If $\frac{P(\mathbf{R'};\alpha)}{P(\mathbf{R};\alpha)} > 1$, then $\mathbf{R'}$ is the new position and the new step is accepted.

 2. $r \leq \frac{P(\mathbf{R'};\alpha)}{P(\mathbf{R};\alpha)}$ where $r$ is a random number generated with uniform PDF such that $ r \epsilon [0,1]$, then this step is also accepted.

 **Note:** In the ising model or classical monte carlo, we were flipping one spin at a time. In here, we will be changing the position of the particle to a trial positon $\mathbf{R'}$ by random and then we evaluate the ratio between the two probabilities. **We don't need to calculate the norm of the wavefunction as it is an impossible task and also we are just computing the ratio between the probabilities.**

 While writing the algorithm of the VMC, one should be prepared with the formulae for the local energy $E_L$ and the wavefunction needed in order to compute the ratios of probabilities in the metropolis algorithm. 

These two functions are almost called as a **random number generator.**

## The Detailed Algorithm:

1. **Initialisation:** Fix the number of Monte-Carlo steps and thermalization steps. Choose an initial $\mathbf{R}$ and variational parameters $\alpha$ and calculate $|\Psi_T(\mathbf{R};\alpha)|^2$. Define also the value of the stepsize to be used when moving from one value of $\mathbf{R}$ to a new one.

2. Initialise the energy and variance.

3. Start the Monte-Carlo calculation with a loop over a given number of Monte-Carlo cycles:

a. Calculate a trial positon $\mathbf{R_p} = \mathbf{R + r\Delta R}$ where $r$ is a random variable $r \epsilon [0,1]$

b. Use then the metropolis algorithm to accept or reject this move by calculating the ratio $ w = \frac{P(R_p)}{P(R)}$, if $w\geq s$ where s is a random number $s\epsilon[0,1]$, the new position is accepted, else we stay at the same place.

c. If the step is accepted, then we say $R = R_p$

d. Update the local energy and the variance

When the Monte Carlo sampling is finished, we calculate the mean energy and the standard devia
tion. Finally, we may print our results to a specified file.

# 1D Quantum Harmonic Oscillator:

## First Illustration of Variational Monte Carlo Methods

The harmonic oscillator in one dimension lends itself nicely for illustrative purposes. The Hamiltonian is

$$
H = -\frac{\hbar^2}{2m} \frac{d^2}{dx^2} + \frac{1}{2} k x^2,
$$

where \( m \) is the mass of the particle and \( k \) is the force constant, e.g., the spring tension for a classical oscillator. In this example we will make life simple and choose \( m = \hbar = k = 1 \). We can rewrite the above equation as

$$
H = -\frac{d^2}{dx^2} + x^2.
$$

The energy of the ground state is then \( E_0 = 1 \). The exact wave function for the ground state is

$$
\Psi_0(x) = \frac{1}{\pi^{1/4}} e^{-x^2/2},
$$

but since we wish to illustrate the use of Monte Carlo methods, we choose the trial function

$$
\Psi_T(x) = \frac{\sqrt{\alpha}}{\pi^{1/4}} e^{-x^2 \alpha^2 / 2}.
$$

Inserting this function in the expression for the local energy in Eq. (14.3), we obtain the following **expression for the local energy**

$$
E_L(x) = \alpha^2 + x^2 (1 - \alpha^4),
$$

with the expectation value for the Hamiltonian of Eq. (14.4) given by

$$
\langle E_L \rangle = \int_{-\infty}^{\infty} |\psi_T(x)|^2 E_L(x) dx,
$$

which reads with the above trial wave function

$$
\langle E_L \rangle =
\frac{\int_{-\infty}^{\infty} dx e^{-x^2 \alpha^2} \left( \alpha^2 + x^2 (1 - \alpha^4) \right)}
{\int_{-\infty}^{\infty} dx e^{-x^2 \alpha^2}}.
$$

Using the fact that

$$
\int_{-\infty}^{\infty} dx e^{-x^2 \alpha^2} = \frac{\sqrt{\pi}}{\alpha^2},
$$

we obtain

$$
\langle E_L \rangle = \frac{\alpha^2}{2} + \frac{1}{2\alpha^2}.
$$

and the variance

$$
\sigma^2 = \frac{(\alpha^4 - 1)^2}{2\alpha^4}.
$$


# The Code:

```python

#load the required modules
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import warnings
```

```python
#defining the wavefunction
def QHO_GS1D(x, alpha=1):
    """
    This function defines the first wavefunction of the quantum harmonic oscillator
    
    Parameters:
    - x: The position of the particle in the 1D QHO
    - alpha: Optimizing parameter, initiate with some value
             (By default = 1)
             
    Returns:
    
    This function returns you the value of the wavefunction for a particular position and optimizing parameter alpha
    """
    
    if not isinstance(x,(float,int)):
        raise ValueError("Value other than float or integer was called")
    if not isinstance(alpha, (float, int)):
        raise ValueError("Value other than float or integer was called")
    
    return np.sqrt(alpha) * np.exp(- (x**2) * (alpha**2) / 2) / (np.pi ** (1/4))
```

```python
#defining the probability density function
def GS_PDF(x,alpha=2):
    """
    This function defines the PDF of the 
    ground state wavefunction of QHO
    
    Parameters:
    - x: The position of the particle in the 1D QHO
    - alpha: Optimizing parameter, initiate with some value
             (By default = 1)
    
    Returns:
    
    This function returns you the value of the PDF for a particular position and optimizing parameter alpha
    corresponding to the ground state of the Quantum Harmonic Oscillator
    """
    
    if not isinstance(x,(float,int)):
        raise ValueError("Value other than float or integer was called")
    if not isinstance(alpha, (float, int)):
        raise ValueError("Value other than float or integer was called")
    
    wave_func = QHO_GS1D(x, alpha)
    return np.abs(wave_func)**2
```

```python
#defining the local energy function
def local_energy(x, alpha=2):
    """
    This function defines the local energy of the 
    ground state wavefunction of QHO.

    Parameters:
    - x: The position of the particle in the 1D QHO
    - alpha: Optimizing parameter, initiate with some value
             (By default = 1)

    Returns:
    - Local energy of the ground state of Quantum harmonic oscillator system at position x
    """
    
    if not isinstance(x,(float,int)):
        raise ValueError("Value other than float or integer was called")
    if not isinstance(alpha, (float, int)):
        raise ValueError("Value other than float or integer was called")
    
    return alpha**2 + (x**2) * (1 - alpha**4)
```

```python
#defining the variance of the energy function
def energy_var(alpha=2):
    """
    This function defines the variance in 
    the local energy of the ground state wavefunction of QHO
    
    Parameters:
    - alpha: Optimizing parameter, initiate with some value
             (By default = 1)
    
    Returns:
    - Variance in the Local energy of the ground state of Quantum harmonic oscillator system at position x
    """
    
    return ((alpha**4 - 1)**2) / (2 * alpha**4)  
```

```python
def VMC_sweeps(x, step, samples = 10000,alpha=2):
    """
    This funciton performs Variational Monte Carlo sweeps
    
    Parameters:
    - x: initial position
    - step: step size
    - alpha_list: in takes a list of alpha values for optimization
    - samples: no. of VMC sweeps to be performed
               (By default: 10000)
    """
    position_saved = []
    energy_saved = []
    
    for n in range(samples):
        r = np.random.rand()
        x_new = x + (r-0.5)*step
        
        P_old = GS_PDF(x, alpha)
        P_new = GS_PDF(x_new, alpha)
        ratio = P_new/(P_old + 1e-10)
        
        s = np.random.rand()
        
        if ratio > s:
            x = x_new  
        
        position_saved.append(x)
        energy_saved.append(local_energy(x, alpha))
    
    return position_saved, energy_saved
```

```python
#defining a function to optimize alpha
def alpha_optimizer(alpha_list, x, step, samples=10000):
    """
    This function will optimize the value of alpha for the wavefunction
    
    Parameters:
    - alpha_list: takes in a list of alpha for optimization
    - x: give an initial value of the position for the particle
    - step: give a step size for movement
    - samples: enter the number of VMC sweeps you would like to perform
               (By default: 10000)
    """
    saved_energies = []
    variance = []
    for a in tqdm(alpha_list, unit='alpha', desc='Optimizing alpha'):
        positions, energies = VMC_sweeps(x, step, samples, a) #energies gives a list of energy values for a particular round, so you need to mean it up
        saved_energies.append(np.mean(energies))
        variance.append(np.var(energies))

    optimal_alpha = alpha_list[np.argmin(saved_energies)]
    
    return saved_energies, optimal_alpha, variance
```

```python
a = np.linspace(0.5,1.5,11)
energies, optimal_alpha, variance = alpha_optimizer(a, 0, 1,100000)
```

```python
z = (a**2)/2 + 1/(2*a**2)
```

```python
plt.plot(a, energies, 'k', linestyle = '-', marker = '+', label='VMC sampled energies')
plt.axvline(optimal_alpha, color = 'r', linestyle='--', label=r'Optimal value of $\alpha$')
plt.axhline(1, linestyle='--', label='Expected energy')
plt.plot(a,z, linestyle = '-.', label='Expected Energy curve')
plt.grid()
plt.legend()
plt.xlabel(r"$\alpha$")
plt.ylabel(r"Energy of the state")
plt.title(r"Energy v/s $\alpha$ plot for a quantum harmonic oscillator")
```

```python
#using optimized alpha to recover the PDF of the system
pos, ene = VMC_sweeps(x=0, step=1, samples=100000, alpha=optimal_alpha)
```

```python
pos = np.array(pos) 
x_vals = np.linspace(min(pos), max(pos), 100)
psi_analytic = np.sqrt(optimal_alpha) * np.exp(- (x_vals**2) * (optimal_alpha**2) / 2) / (np.pi ** (1/4))
#Normalization of the analytic wavefunction
psi_sq_analytic = psi_analytic**2 #Un-normalized PDF
psi_sq_analytic /= np.trapz(psi_sq_analytic, x_vals) #Normalized PDF


plt.figure(figsize=(8,5))
plt.plot(x_vals, psi_sq_analytic, label="Analytic Probability Density", linestyle="--", color='k')


plt.hist(pos, bins=50, color='y', density=True, alpha=0.5, label="VMC Sampled Positions")

plt.xlabel("Position (x)")
plt.ylabel("Probability Density of QHO")
plt.legend()
plt.grid()
plt.title("Comparison of VMC Sampling and Analytic Solution")
```

```python
plt.hist(ene, bins=50, density=True, alpha=0.5, color='k', label="VMC Sampled Energies")
plt.xlabel("Energy")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
print("The mean energy of the particle in the ground state of Quantum harmonic oscillator is:", np.mean(ene))
```

# Hydrogen atom

```python
#defining the trial wavefunction for the ground state of hydrogen atom
def Hyd_GS(r, alpha = 2):
    """
    This function defines the Ground state wavefunction of Hydrogen atom    
    Parameters:
    - r: The position of the electron in the hydrogen atom
    - alpha: Optimizing parameter, initiate with some value
            (By default = 2)
             
    Returns:
    
    This function returns you the value of the wavefunction for a particular position and optimizing parameter alpha
    """
    
    if not isinstance(r,(float,int)):
        raise ValueError("Value other than float or integer was called")
    if not isinstance(alpha, (float, int)):
        raise ValueError("Value other than float or integer was called")
    
    return alpha * r * np.exp(-alpha*r)
```

```python
def Hyd_GSPDF(r, alpha = 1):
    """
    This function defines the PDF of the 
    ground state wavefunction of Hydrogen atom
    
    Parameters:
    - r: The position of the particle in the Hydrogen atom 
    - alpha: Optimizing parameter, initiate with some value
             (By default = 1)
    
    Returns:
    
    This function returns you the value of the PDF for a particular position and optimizing parameter alpha
    corresponding to the ground state of the Hydrogen atom
    """
    
    if not isinstance(r,(float,int)):
        raise ValueError("Value other than float or integer was called")
    if not isinstance(alpha, (float, int)):
        raise ValueError("Value other than float or integer was called")
    
    wave_func = Hyd_GS(r, alpha)
    return np.abs(wave_func)**2
```

```python
#defining the local energy function
def Hyd_local(r, alpha=2):
    """
    This function defines the local energy of the 
    ground state wavefunction of Hydrogen atom.

    Parameters:
    - r: The position of the particle in the Hydrogen atom
    - alpha: Optimizing parameter, initiate with some value
             (By default = 2)

    Returns:
    - Local energy of the ground state of Quantum harmonic oscillator system at position x
    """
    
    if not isinstance(r,(float,int)):
        raise ValueError("Value other than float or integer was called")
    if not isinstance(alpha, (float, int)):
        raise ValueError("Value other than float or integer was called")
    
    return - 1/r - (alpha/2) * (alpha - (2/r))
```

```python
def Hyd_VMC(r, step, samples = 10000,alpha=2):
    """
    This funciton performs Variational Monte Carlo sweeps
    
    Parameters:
    - r: initial position
    - step: step size
    - alpha_list: in takes a list of alpha values for optimization
    - samples: no. of VMC sweeps to be performed
               (By default: 10000)
    """
    position_saved = []
    energy_saved = []
    
    for n in range(samples):
        q = np.random.uniform(-step, step)
        r_new = r + q
        if r_new <= 0:
            continue
        
        P_old = Hyd_GSPDF(r, alpha)
        P_new = Hyd_GSPDF(r_new, alpha)
        ratio = P_new/(P_old + 1e-10)
        
        s = np.random.rand()
        
        if ratio > s:
            r = r_new  
        
        position_saved.append(r)
        energy_saved.append(Hyd_local(r, alpha))
    
    return position_saved, energy_saved
```

```python
#defining a function to optimize alpha
def Hyd_alpha_opt(alpha_list, r, step, samples=10000):
    """
    This function will optimize the value of alpha for the wavefunction
    
    Parameters:
    - alpha_list: takes in a list of alpha for optimization
    - r: give an initial value of the position for the particle
    - step: give a step size for movement
    - samples: enter the number of VMC sweeps you would like to perform
               (By default: 10000)
    """
    saved_energies = []
    variance = []
    for a in tqdm(alpha_list, unit='alpha', desc='Optimizing alpha'):
        positions, energies = Hyd_VMC(r, step, samples, a) #energies gives a list of energy values for a particular round, so you need to mean it up
        saved_energies.append(np.mean(energies))
        variance.append(np.var(energies))

    optimal_alpha = alpha_list[np.argmin(saved_energies)]
    
    return saved_energies, optimal_alpha, variance
```

```python
a = np.linspace(0.5,1.5,11)
energies, optimal_alpha, variance = Hyd_alpha_opt(a, 0.5, 25,100000)
```
